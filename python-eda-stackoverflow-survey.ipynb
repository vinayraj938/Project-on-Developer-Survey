Convert the Series to DataFrame:

python

python

Creates an empty list options to keep track of the unique values (columns) that will be added to the DataFrame.

python

Loops through each non-null entry in col_series. idx is the index, and value is the entry at that index.

Checks if the option is already a column in result_df. If not, it adds the option to the list options and creates a new column in result_df with all values set to False.

Sets the value at the specified idx and option column to True, indicating that this option is present in the original entry at index idx.

Returns the DataFrame result_df, but only the columns specified in options, which are the unique values extracted from the original Series.
Usage Example
python
Copy code
# Assuming survey_df is defined and contains the 'DevType' column
result = split_multicolumn(survey_df['DevType'])
This line calls the function split_multicolumn with the DevType column from the survey_df DataFrame. The result will be a new DataFrame with each unique DevType value as a column, and True/False values indicating whether each entry contains that DevType.

```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame() # coverting to pandas readabale format
    options = [] # creating a empty list adding the values
    # Iterate over the column
    for idx, value in col_series[col_series.notnull()].items():#  index and values are selected by the loop
        # Break each value into list of options
        for option in value.split(';'): # splits each value in the data frame
            option = option.strip()  # Remove leading and trailing spaces
            # Add the option as a column to result
            if option not in result_df.columns: # if the option is not in the column in a specified index make it flase
                options.append(option)
                result_df[option] = False # if option is not present in the columns making it to false
            # Mark the value in the option column as True
            result_df.at[idx, option] = True # if option exist in the column making it to true
    return result_df[options]

# Assuming survey_df is defined and contains the 'DevType' column
result = split_multicolumn(survey_df['DevType'])


```

#### Q:  Which are the most loved languages, i.e., a high percentage of people who have used the language want to continue learning & using it over the next year?

While this question may seem tricky at first, it's straightforward to solve using Pandas array operations. Here's what we can do:

- Create a new data frame `languages_loved_df` that contains a `True` value for a language only if the corresponding values in `languages_worked_df` and `languages_interested_df` are both `True`
- Take the column-wise sum of `languages_loved_df` and divide it by the column-wise sum of `languages_worked_df` to get the percentage of respondents who "love" the language
- Sort the results in decreasing order and plot a horizontal bar graph

```python
language_loved_df = language_worked_df & language_interested_df
language_loved_df
```

```python
language_loved_percentage = (language_loved_df.sum() * 100 /language_worked_df.sum() ) .sort_values(ascending=False)
```

```python
language_loved_percentage
```

```python
plt.figure(figsize=(12, 12))
sns.barplot(x=language_loved_percentage, y=language_loved_percentage.index)
plt.title("Most loved languages");
plt.xlabel('count');
```
